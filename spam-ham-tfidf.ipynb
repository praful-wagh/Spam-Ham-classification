```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split as tts
import re
from tqdm import tqdm
from gensim.models import Word2Vec
```

```python
df = pd.read_csv('/kaggle/input/spam-ham-dataset/spam_dataset.csv')
```

```python
df = df.drop(df.columns[0:2], axis=1)
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
x = df.text
x.head()
```

```python
x.shape
```

```python
y = df.label_num
y.head()
```

```python
y.shape
```

```python
import nltk
nltk.download('wordnet')
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()
ps = PorterStemmer()
from nltk.corpus import stopwords
stopwords= stopwords.words('english')
```

```python
nltk.download()
```

```python
corpus = []
for i in tqdm(x):
    r = (re.sub('[^a-zA-Z]',' ',i)).lower().split()
    r = [lem.lemmatize(w) for w in r if not w in stopwords]
    corpus.append(' '.join(r))
```

```python
x = corpus
x[:2]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
x = TfidfVectorizer().fit_transform(x)
```

```python
xtr, xte, ytr, yte = tts(x,y,test_size=.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
xtr.shape,len(ytr)
```

```python
cls = LogisticRegression()
cls.fit(xtr,ytr)
yp = cls.predict(xte)
accuracy_score(yte, yp)
```

```python
from sklearn.naive_bayes import GaussianNB
cls = GaussianNB()
cls.fit(xtr.todense(),ytr)
yp = cls.predict(xte.todense())
accuracy_score(yte, yp)
```
